```python
# 2. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен DAU 
# *DAU (Daily Active Users) — это метрика, которая показывает количество уникальных пользователей, которые взаимодействовали с продуктом, 
# приложением или сервисом хотя бы один раз в течение дня. DAU помогает понять, сколько пользователей активно пользуются 
# продуктом каждый день.
```

```python
import pandas as pd

# Загружаем данные
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Данные об аудитории")

# Преобразуем дату
df["date"] = pd.to_datetime(df["date"], errors='coerce')

# Отбираем только ноябрь
nov_df = df[df["date"].dt.month == 11]

# Создаем колонку с датой без времени
nov_df["only_date"] = nov_df["date"].dt.date

# Считаем уникальных пользователей за каждый день
dau_per_day = nov_df.groupby("only_date")["user_id"].nunique()

# Среднее DAU за ноябрь
DAU = int(dau_per_day.mean())

print("Средний DAU за ноябрь:", DAU)
```

```python
# 3. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен retention первого дня у пользователей, пришедших в продукт 1 ноября 
# *Retention (удержание пользователей) — это метрика, которая показывает, сколько пользователей продолжает пользоваться 
# продуктом через определенный промежуток времени после первоначального взаимодействия. Retention можно рассчитать как процент пользователей, 
# вернувшихся в продукт через определенное время (например, через 1 день, 1 неделю, 1 месяц) от количества всех новых пользователей.
```

```python
import pandas as pd

# Загружаем данные
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Данные об аудитории")

# Преобразуем дату
df["date"] = pd.to_datetime(df["date"], errors='coerce')
df["only_date"] = df["date"].dt.date

# 1. Пользователи, пришедшие впервые 1 ноября
nov1_users = df[df["only_date"] == pd.to_datetime("2023-11-01").date()]["user_id"].unique()

# 2. Пользователи, вернувшиеся 2 ноября
nov2_users = df[df["only_date"] == pd.to_datetime("2023-11-02").date()]["user_id"].unique()

# 3. Находим пересечение (вернувшиеся)
returned_users = set(nov1_users).intersection(set(nov2_users))

# 4. Считаем retention первого дня
retention_1day = len(returned_users) / len(nov1_users) * 100

print(f"Retention первого дня: {retention_1day:.1f}%")
```

```python
# 5. Во вкладке "Данные об аудитории" есть информация о том, сколько объявлений посмотрел каждый пользователь (view_adverts). 
# Посчитайте пользовательскую конверсию в просмотр объявления за ноябрь? (в пользователях) 
# * Пользовательская конверсия — это метрика, которая показывает, какой процент пользователей выполнил целевое действие 
# по отношению к общему количеству пользователей. В контексте веб-сайтов это может быть действие, такое как просмотр объявления 
# или клик по рекламному баннеру. 
```

```python
import pandas as pd

# Загружаем данные
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Данные об аудитории")

# Преобразуем дату
df["date"] = pd.to_datetime(df["date"], errors='coerce')

# Фильтруем пользователей за ноябрь
nov_df = df[df["date"].dt.month == 11]

# Уникальные пользователи за ноябрь
total_users = nov_df["user_id"].nunique()

# Пользователи, которые посмотрели хотя бы одно объявление
converted_users = nov_df[nov_df["view_adverts"] > 0]["user_id"].nunique()

# Конверсия в проценте
conversion = converted_users / total_users * 100

print(f"Пользовательская конверсия в просмотр объявления за ноябрь: {conversion:.1f}%")
```

```python
# 6. Используя информацию из вкладки "Данные об аудитории", посчитайте среднее количество просмотренных объявлений на пользователя в ноябре
# 4,9 6,2 5,3 2,9
```

```python
import pandas as pd

# Загружаем данные
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Данные об аудитории")

# Преобразуем дату
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Оставляем только ноябрь
nov_df = df[df["date"].dt.month == 11]

# Считаем сумму просмотров объявлений
total_views = nov_df["view_adverts"].sum()

# Считаем количество уникальных пользователей
total_users = nov_df["user_id"].nunique()

# Среднее количество просмотров на пользователя
avg_views_per_user = total_views / total_users

print("Сумма просмотров:", total_views)
print("Уникальных пользователей:", total_users)
print(f"Среднее просмотров на пользователя: {avg_views_per_user:.1f}")
```

```python
# 8. Во вкладке "Данные АБ-тестов" результаты трех несвязанных АБ тестов для ARPU (общая выручка/общее количество пользователей).
# Посмотрите на результаты тестов и интерпретируйте их. Напишите значения p-value, которые вы получили.
# Подготовьте выводы и рекомендации. 

# experiment_num - номер эксперимента
# experiment_group - группа, в которую попал пользователь
# user_id - id пользователя
# revenue - выручка, которую сгенерировал пользователь, купив платную услугу продвижения
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Загружаем данные
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Данные АБ тестов")

alpha = 0.05

for exp in sorted(df["experiment_num"].unique()):
    exp_df = df[df["experiment_num"] == exp].copy()
    groups = exp_df["experiment_group"].unique()
    
    g1_name = groups[0]
    g2_name = groups[1]
    
    g1 = exp_df[exp_df["experiment_group"] == g1_name]["revenue"]
    g2 = exp_df[exp_df["experiment_group"] == g2_name]["revenue"]
    
    arpu_1 = g1.mean()
    arpu_2 = g2.mean()
    
    stat, p_value = ttest_ind(g1, g2, equal_var=False)
    
    print(f"\n=== Эксперимент {exp} ===")
    print(f"{g1_name}: ARPU = {arpu_1:.2f}")
    print(f"{g2_name}: ARPU = {arpu_2:.2f}")
    print(f"p-value = {p_value:.5f}")
    
    if p_value < alpha:
        print("Результат статистически значим (p < 0.05)")
        if arpu_2 > arpu_1:
            print(f"Рекомендуется группа {g2_name}, т.к. ARPU выше.")
        else:
            print(f"Рекомендуется группа {g1_name}, т.к. ARPU выше.")
    else:
        print("Статистически значимых различий не обнаружено (p >= 0.05)")
        print("Рекомендуется не выкатывать изменения без дополнительных данных.")
```

```python
import pandas as pd

# Загружаем лист "Листеры"
df = pd.read_excel("Данные для тестового задания.xlsx", sheet_name="Листеры")

# Посмотрим первые строки
print(df.head())

# Общая выручка
total_revenue = df["revenue"].sum()

# Количество уникальных пользователей
total_users = df["user_id"].nunique()

# Средний доход на пользователя (ARPU)
arpu = total_revenue / total_users

print("Общая выручка:", total_revenue)
print("Уникальных пользователей:", total_users)
print(f"Средний доход на пользователя (ARPU): {arpu:.1f}")
```
